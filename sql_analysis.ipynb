```python
import pandas as pd
import sqlite3

df = pd.read_csv('C:/Users/CHANDRIMA/Desktop/superstore-analysis/data/superstore_cleaned.csv')
conn = sqlite3.connect('C:/Users/CHANDRIMA/Desktop/superstore-analysis/data/superstore.db')
df.to_sql('sales', conn, if_exists='replace', index=False)
print("Database created successfully!")
print(f"Total rows loaded: {len(df)}")
```

```python
q1 = pd.read_sql("""
    SELECT Region, 
           ROUND(SUM(Sales), 2) AS Total_Sales,
           ROUND(SUM(Profit), 2) AS Total_Profit
    FROM sales
    GROUP BY Region
    ORDER BY Total_Profit DESC
""", conn)
print("QUERY 1 - Sales and Profit by Region:")
print(q1)
```

```python
q2 = pd.read_sql("""
    SELECT Category, "Sub-Category",
           ROUND(SUM(Profit), 2) AS Total_Profit,
           ROUND(AVG("Profit Margin %"), 2) AS Avg_Margin
    FROM sales
    GROUP BY Category, "Sub-Category"
    ORDER BY Total_Profit ASC
    LIMIT 10
""", conn)
print("QUERY 2 - Most Loss-Making Product Sub-Categories:")
print(q2)
```

```python
q3 = pd.read_sql("""
    SELECT "Customer Name",
           ROUND(SUM(Sales), 2) AS Total_Revenue,
           COUNT(DISTINCT "Order ID") AS Total_Orders
    FROM sales
    GROUP BY "Customer Name"
    ORDER BY Total_Revenue DESC
    LIMIT 10
""", conn)
print("QUERY 3 - Top 10 Customers by Revenue:")
print(q3)
```

```python
q4 = pd.read_sql("""
    SELECT "Order Year",
           ROUND(SUM(Sales), 2) AS Annual_Sales,
           ROUND(SUM(Profit), 2) AS Annual_Profit
    FROM sales
    GROUP BY "Order Year"
    ORDER BY "Order Year"
""", conn)
print("QUERY 4 - Year Over Year Sales Trend:")
print(q4)
```

```python
q5 = pd.read_sql("""
    SELECT "Ship Mode",
           ROUND(AVG("Delivery Days"), 1) AS Avg_Delivery_Days,
           COUNT(*) AS Total_Orders
    FROM sales
    GROUP BY "Ship Mode"
    ORDER BY Avg_Delivery_Days
""", conn)
print("QUERY 5 - Average Delivery Days by Ship Mode:")
print(q5)
```
